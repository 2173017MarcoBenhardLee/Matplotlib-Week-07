```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

3D Axes

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

plt.show()
```

3D Line Plot

```python
z = np.linspace(0, 1, 100) # nilai terkecil 0 terbesar 1 sebanyak 100 bilangan
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
fig = plt.figure()

ax = plt.axes(projection='3d')

ax.plot(x, y, z, color='red')

ax.set(title='Contoh 3D Line Plot',
      xlabel='Sumbu X',
      ylabel='Sumbu Y',
      zlabel='Sumbu Z')

plt.show()
```

```python
3D Scatter plot
```

```python
z = np.linspace(0, 1, 100) # nilai terkecil 0 terbesar 1 sebanyak 100 bilangan
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

fig = plt.figure()

ax = plt.axes(projection='3d')

ax.scatter(x, y, z, color='green')

ax.set(title='Contoh 3D Scatter Plot',
      xlabel='Sumbu X',
      ylabel='Sumbu Y',
      zlabel='Sumbu Z')

plt.show()
```
